## If we go to the dream11 point system we see a player's point gets doubled if the player is captain of the match, let's figure out if a player is captain. unfortunately the player list doesn't have the information about the captain, its located on another table player_captain, so we'll merge these two table to extract the captain information. Unfortunately we don't have any information about the vice captain, so we are not gonna calculates the points for vice captain

```python
player.merge(player_captain,on='Player_Id')
```

## Let's get the usefull columns

```python
player.merge(player_captain,on='Player_Id')[['Player_Name','Match_Id','Is_Captain']]
```

## Store it in a variable

```python
captain = player.merge(player_captain,on='Player_Id')[['Player_Name','Match_Id','Is_Captain']]
```

```python
delivery.head(1)
```

```python
captain.head(1)
```

## Now we'll merge this new dataframe with our delivery dataframe, we'll do left_merge on match_id and batter, and right_merge on Match_Id and Player_Name

```python
delivery.merge(captain,left_on=['match_id','batter'],right_on=['Match_Id','Player_Name'])
```

```python
captain.shape
```

```python
delivery.shape
```

## So problem with this approach is, we didn't provide value to 'how' parameter so pandas perform inner join inside because default value of 'how' parameter inner, and as we know the inner join only performs between the common values so it looses so many values as we can see it only include around 16000 values it might ber bacause there are so many matches that presented on delivery dataframe but didn't present on the player dataframe, may be because of the mismatched in the dataset.

## So we'll perform left join to include all the values from delivery dataframe.

```python
delivery.merge(captain,left_on=['match_id','batter'],right_on=['Match_Id','Player_Name'],how='left')
```

## We can handle this problem with filling the nan values in is_captain column with 0, means the player wasn't the captain

```python
delivery.merge(captain,left_on=['match_id','batter'],right_on=['Match_Id','Player_Name'],how='left').fillna(0)
```

```python
delivery = delivery.merge(captain,left_on=['match_id','batter'],right_on=['Match_Id','Player_Name'],how='left').fillna(0)
```

## So as we can see we need the total number of runs and the numbers of ball played are needed to calculate the strike rate and other points
## Let's first calculate the total runs by each player

```python
delivery.groupby(['match_id','batter'])['batsman_runs'].sum().reset_index()
```

```python
runs = delivery.groupby(['match_id','batter'])['batsman_runs'].sum().reset_index()
```

## For the total balls played we'll do somrthing like this

```python
delivery.groupby(['match_id','batter'])['ball'].count().reset_index()
```

```python
balls = delivery.groupby(['match_id','batter'])['ball'].count().reset_index()
```

## Now to calculate how many 4s

```python
delivery.query('batsman_runs==4')
```

```python
delivery.query('batsman_runs==4').groupby(['match_id','batter'])['batsman_runs'].count().reset_index()
```

```python
fours = delivery.query('batsman_runs==4').groupby(['match_id','batter'])['batsman_runs'].count().reset_index()
```

## For sixes

```python
delivery.query('batsman_runs==6').groupby(['match_id','batter'])['batsman_runs'].count().reset_index()
```

```python
sixes = delivery.query('batsman_runs==6').groupby(['match_id','batter'])['batsman_runs'].count().reset_index()
```

## Now we'll merge these four dataframes runs, balls, fours, sixes together

```python
runs.head(1)
```

```python
runs.shape
```

```python
balls.head(1)
```

```python
balls.shape
```

## Now let's runs and balls together to get the total number of balls played and run scored, and it's gonna be a inner join as both the dataframe has same number of rows, and we need all the values

```python
runs.merge(balls,on=['match_id','batter'])
```

## Now we'll merge fours and sixes with this new merge with left join so we don't loose any rows, because every balls might not scored a four or a six

```python
fours.head(1)
```

```python
fours.shape
```

## Let's merge fours first, it's gonna be a left join

```python
runs.merge(balls,on=['match_id','batter']).merge(fours,on=['match_id','batter'],how='left')
```

## now let's merge sixes, also it's gonna be a left join

```python
runs.merge(balls,on=['match_id','batter']).merge(fours,on=['match_id','batter'],how='left').merge(sixes,on=['match_id','batter'],how='left')
```

```python
score_df = runs.merge(balls,on=['match_id','batter']).merge(fours,on=['match_id','batter'],how='left').merge(sixes,on=['match_id','batter'],how='left')
```

## We need to rename the columns

```python
score_df.rename(columns={'batsman_runs_x':'runs','ball':'balls','batsman_runs_y':'fours','batsman_runs':'sixes'})
```

## Let's replace the nan values with 0, here nan means no fours or sixes for those batsmans

```python
score_df.rename(columns={'batsman_runs_x':'runs','ball':'balls','batsman_runs_y':'fours','batsman_runs':'sixes'}).fillna(0)
```

```python
score_df = score_df.rename(columns={'batsman_runs_x':'runs','ball':'balls','batsman_runs_y':'fours','batsman_runs':'sixes'}).fillna(0)
```

## Now let's calculate the strike rate
